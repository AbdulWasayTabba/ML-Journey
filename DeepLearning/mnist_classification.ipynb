```python
import tensorflow
from tensorflow import keras
from tensorflow.keras import Sequential
from tensorflow.keras.layers import Dense,Flatten
```

```python



(X_train,y_train),(X_test,y_test) = keras.datasets.mnist.load_data()
```

```python
# we have to bring the value between range 0 and 1
# because cause of this weights are more accurate
# and convergence is fast
# we will do that by dividing the numbers by max which is 255
# 255/255 = 1 and  0 / 255 = 0
```

```python
X_train = X_train/255
X_test = X_test/255
```

```python
X_train
```

```python
model = Sequential()
model.add(Flatten(input_shape=(28,28)))#this convert it do 1d(1x784) from (28x28)
model.add(Dense(128,activation='relu'))
model.add(Dense(32,activation='relu'))
model.add(Dense(10,activation='softmax'))#when more than 1 node in output use softmax
```

```python
model.summary()
```

```python
model.compile(loss='sparse_categorical_crossentropy',optimizer='Adam',metrics=['accuracy'])
```

```python
history = model.fit(X_train,y_train,epochs=25,validation_split=0.2)
```

```python
y_prob =model.predict(X_test)
```

```python
y_prob
```

```python
y_pred = y_prob.argmax(axis=1)
```

```python
y_pred
```

```python
from sklearn.metrics import accuracy_score

accuracy_score(y_test,y_pred)
```

```python
plt.plot(history.history['loss'])
plt.plot(history.history['val_loss'])
```

```python
plt.plot(history.history['accuracy'])
plt.plot(history.history['val_accuracy'])
```
