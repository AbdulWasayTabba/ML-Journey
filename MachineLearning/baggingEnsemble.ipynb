```python
import numpy as np
import pandas as pd
from sklearn.datasets import make_classification
from sklearn.ensemble import BaggingClassifier,RandomForestClassifier
from sklearn.tree import plot_tree
from sklearn.tree import DecisionTreeClassifier
from mlxtend.plotting import plot_decision_regions
from sklearn.metrics import accuracy_score
import matplotlib.pyplot as plt

# Load the dataset
# You requested the exact code, so this path is unchanged. 
# If running locally, you must have the file at this specific location or update this line manually.
df = pd.read_csv('C:\\Users\\HP\\Desktop\\datascienceNumpy\\MachineLearning\\Iris.csv')

# Preprocessing
df = df.iloc[:,1:]
X = df.iloc[:,0:2]
y = df.iloc[:,-1]

# Taking only 100 rows for training to simulate the "Learning Tool" environment
df = df.sample(100)
df_train = df.iloc[:60,:].sample(10)
df_val = df.iloc[60:80,:].sample(5)
df_test = df.iloc[80:,:].sample(5)

X_test = df_val.iloc[:,0:2].values
y_test = df_val.iloc[:,-1].values

# Function to evaluate and plot the decision tree and regions
def evaluate(clf,X,y):
    clf.fit(X,y)
    plot_tree(clf)
    plt.show()
    plot_decision_regions(X.iloc[:,0:2].values,y.values,clf=clf,legend=2)
    y_pred = clf.predict(X_test)
    print(accuracy_score(y_test,y_pred))

# Case 1 - Bagging (Tree 1)
# Sampling with replacement
df_bag = df_train.sample(8,replace=True)
X = df_bag.iloc[:,0:2]
y = df_bag.iloc[:,-1]

print("Tree 1")
dt_bag1 = DecisionTreeClassifier()
evaluate(dt_bag1,X,y)

# Case 1 - Bagging (Tree 2)
# Sampling with replacement again
df_bag = df_train.sample(8,replace=True)
X = df_bag.iloc[:,0:2]
y = df_bag.iloc[:,-1]

print("Tree 2")
dt_bag2 = DecisionTreeClassifier()
evaluate(dt_bag2,X,y)

# Case 1 - Bagging (Tree 3)
# Sampling with replacement again
df_bag = df_train.sample(8,replace=True)
X = df_bag.iloc[:,0:2]
y = df_bag.iloc[:,-1]

print("Tree 3")
dt_bag3 = DecisionTreeClassifier()
evaluate(dt_bag3,X,y)

# Aggregation Logic (Manual Prediction for a sample point)
print("Prediction Tree 1:", dt_bag1.predict(X_test[0].reshape(1,-1)))
print("Prediction Tree 2:", dt_bag2.predict(X_test[0].reshape(1,-1)))
print("Prediction Tree 3:", dt_bag3.predict(X_test[0].reshape(1,-1)))

# Pasting (Sampling WITHOUT replacement)
# This is usually the second part of the notebook
print("Pasting Example (No Replacement)")
df_bag = df_train.sample(8,replace=False)
X = df_bag.iloc[:,0:2]
y = df_bag.iloc[:,-1]

dt_pasting = DecisionTreeClassifier()
evaluate(dt_pasting,X,y)
```
